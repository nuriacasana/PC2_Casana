```python
import pandas as pd
datos_raw = 'https://github.com/nuriacasana/PC2-NURIA/raw/refs/heads/main/HDI.xlsx'

# cargado de archivo excel
humanData = pd.read_excel(datos_raw)

# Mostrarnos la data del excel
humanData
```

```python
# Conocer el tipo de dato
humanData.info()
```

```python
# Primer paso es realizar la eliminación de filas que no tengas datos y que no sean datos necesarios

# Nombre de la variable columnas
humanData1= humanData.columns

# De manera manual haremos la selección de columnas que no necesitamos
drop_indices=[3, 5, 7, 9, 11,13,14]
drop_columns= [ humanData1[ i] for i in drop_indices if i< len(humanData1)]

# Eliminación
humanData= humanData.drop(columns=drop_columns)

names_column= ["Rank","Country", "Human_Development_Index", "Life_expectancy_at_birth" , "Expected_years_of_schooling", "Mean_years_of_schooling", "Gross_national_income_per_capita", "GNI_per_capita_rank_minus_HDI_rank"]

# Asignacion de los nuevos nombres de las columnas

humanData.columns= names_column

humanData

```

```python
# Eliminación de las filas que no tienen data
# se visualizó que del 0-4 habian datos innecesarios

humanData= humanData.drop(range(5))

humanData
```

```python
#Filtrar y mostrar las filas del DataFrame donde la columna 'Rank' tiene valores faltantes (NaN)
filas_sin_rank = humanData[humanData['Rank'].isnull()]
filas_sin_rank
# Visualizar los rangos para las categorias del desarrollo humano
```

```python
# eliminar filas vacías

# Eliminar filas con valores NaN en cualquier columna
humanData = humanData.dropna()

humanData
```

```python
# Eliminamos filas_sin_rank, guardando los intervalos de desarrollo humano
# subset:Indica que solo debe considerar la columna 'Rank' al momento de decidir si una fila debe eliminarse.
humanData= humanData.dropna( subset=[ 'Rank'])
humanData
```

```python
#Eliminar la columna 'Rank', no me es necesario porque ya obtuve los rangos
humanData = humanData.drop(columns=['Rank'])

humanData
```

```python
# Añado columna de categorias por intervalos que antes se guardó
humanData['Categories'] = ''
# Asignación de rangos de intervalo
humanData.loc[7:72, 'Categories' ]= 'VERY HIGH HUMAN DEVELOPMENT'
humanData.loc[74:122, 'Categories' ]= 'HIGH HUMAN DEVELOPMENT'
humanData.loc[124:167, 'Categories' ]= 'MEDIUM HUMAN DEVELOPMENT'
humanData.loc[169:, 'Categories' ]= 'LOW HUMAN DEVELOPMENT'

# reseteamos el index para que el conteo se ordene

humanData.reset_index(drop=True, inplace=True)
humanData
# Muestro tabla con los datos limpios y ordenados
```
